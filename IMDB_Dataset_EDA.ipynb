2. How many movies released in a year?

* we can change the format of date-time in dataset so,we can import libraries of datetime.

```python
from datetime import datetime
from datetime import date
```

```python
df['release_date']=pd.to_datetime(df['release_date'])
```

```python
df["year"]=df["release_date"].apply(lambda x: x.strftime("%y"))
yearly_movies_release=df.groupby("year")["id"].count()
yearly_movies_release
```

```python
fig=plt.figure()
fig.set_figheight(5)
fig.set_figwidth(30)
plt.plot(yearly_movies_release,color="r")
plt.xlabel('years')
plt.ylabel('number of released movies')
plt.title("movies released in a particular year")
```

**INFERENCE**:

year 2013 are released highest numbet of movies trail by 2012,2014.

5. List out the movies and their genres which genres have most common in movies?


```python
list_of_all_genres = []
def extract_genre(list_of_genre):
  for genre in list_of_genre:
    list_of_all_genres.append(genre['name'])
```

```python
def convert_list(l1):
  return eval(l1)
```

```python
df.loc[~df['genres'].isna(), 'genres'] = df.loc[~df['genres'].isna(), 'genres'].apply(convert_list)
```

```python
df_genre = df[~df['genres'].isna()]
df_genre[['title','genres']]
```

```python
df_genre['genres'].apply(extract_genre)
```

```python
set(list_of_all_genres)
```

```python
plt.figure(figsize=(20, 5))
plt.hist(list_of_all_genres)
plt.xlabel('genres')
plt.ylabel('frequency')
plt.title("frequency of genres")
```

INFERENCE:

comedy and drama genres have highest frequency in around 2600 movies.
